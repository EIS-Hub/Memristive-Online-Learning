```python
def pcm_convolve(tensor, nu=0.1, delta_g=1, dt=1, dtype=torch.float32):
    '''
    Given a train of pulses, an eligibility vector is produced, where at each pre-synaptic pulse,
    a PCM device undergoes a progressive set, to then drift following a power-law behavior.
    
    In a first implementation, the parameters of the power law are tuned to maximise performance of
    pcm-prop, later they will be replaced with data calibrated on experimental evidence.
    FM
    '''
    r_shp = range(len(tensor.size()))
    transpose_perm = [1, 0] + list(r_shp)[2:]
    tensor_time_major = tensor.permute(transpose_perm)
    
    ### pcm-trace, non vectorized ###
    filtered_tensor = torch.zeros( tensor.size(1), tensor.size(0), tensor.size(-1) ).type(dtype=dtype)
    eps_ref = torch.zeros( tensor.size(0), tensor.size(-1) ).type(dtype=dtype)
    Tevolve = torch.zeros( tensor.size(0), tensor.size(-1) ).type(dtype=dtype)
    for t in range( tensor.size(1) ):
        Tevolve = (Tevolve - dt*(tensor_time_major[t]-1) )*(tensor_time_major[t]-1)*(-1)
        eps_ref = torch.add( filtered_tensor[t-1].clone(), delta_g )*tensor_time_major[t] + ( eps_ref*(1-tensor_time_major[t]) )
        filtered_tensor[t] = eps_ref * torch.pow( dt/( dt + Tevolve ), nu )
    filtered_tensor = filtered_tensor.permute(transpose_perm)
    return filtered_tensor
```

```python
dtype = torch.float32
```

# PMC-trace optimization
The PCM trace is controlled by two parameters, namely the conductance step $\Delta G$ (when the PCM is progressively SET) and the  decay factor $\nu$. Experimental measurement reveal that the state of the resistance might affect the decay factor $\nu$, while it is well known that the programming conditions influence the conductance step $\Delta G$. 

$\mathcal{F}$ = Exponential-filter ($\tau_{neurons}$)\
$\mathcal{G}$ = Power-Law-filter ($\Delta G$, $\nu$)\
minimize --> $(\mathcal{F}(In) - \mathcal{G}(In))^2$

As a consequence we can play a bit with those parameters to minimize the reconstruction loss that appears when comparing the E-trace (which comes from applying the exponential kernel to the output spike train of the Pre-Neuron) and the PCM-trace (which does the same treatment of e-trace, but with a Power-Law kernel).

The Experiment:
- produce a random (Poisson) train of spikes
- apply the exponential kernel --> target signal
- appply the power law kernel --> input
- compute the MSE
- evaluate the gradient with respect to the DeltaG and nu parameters

```python
batches = 1
neurons = 1
seq_len = 1000

time_step = 1e-3
f_input = 0.2e2 # Hz
p_spike = f_input * time_step

Input_spikes_train = torch.bernoulli( torch.ones(batches,seq_len,neurons)*p_spike ).type(dtype=dtype)
```

```python
# Exponential Filter
tau_F = 20e-3
decay = np.exp(-time_step/tau_F).astype(np.float32)
F = exp_convolve( Input_spikes_train, decay )
```

```python
# PowerLaw Filter
nu = torch.nn.Parameter( torch.tensor(0.1).type(dtype=dtype) )
delta_g = torch.nn.Parameter( torch.tensor((1-decay)/2).type(dtype=dtype) )
Ginit = pcm_convolve( Input_spikes_train, nu=nu, delta_g=delta_g, dt=time_step, dtype=dtype )
```

```python
plt.plot( range(seq_len), F.detach().numpy()[0,:,0], label='Target E-Vector' )
plt.plot( range(seq_len), Ginit.detach().numpy()[0,:,0], label='PCM trace, Init' )
plt.legend()
```

### Training Loop to optimize the PCM trace

```python
iters = 250
params = [nu, delta_g]
optimizer = torch.optim.SGD( params, lr = 1 )
loss_hist = []
for i in range(iters):
    G = pcm_convolve( Input_spikes_train, nu=nu, delta_g=delta_g, dt=time_step )
    MSE = torch.mean( torch.pow( (F-G),2 ) )
    optimizer.zero_grad()
    MSE.backward()
    optimizer.step()
    loss_hist.append( MSE.item() )
```

```python
plt.plot( loss_hist )
```

```python
G_trained = pcm_convolve( Input_spikes_train, nu=nu, delta_g=delta_g, dt=time_step )
```

```python
plt.plot( range(seq_len), F.detach().numpy()[0,:,0], label='Target E-Vector' )
plt.plot( range(seq_len), Ginit.detach().numpy()[0,:,0], label='PCM-trace, init' )
plt.plot( range(seq_len), G_trained.detach().numpy()[0,:,0], label='PCM-trace, trained' )
plt.legend()

print('Optimal set of parameters: nu={}, delta_g={}'.format(nu, delta_g))
```

### Finding a relation between input frequency and (nu, delta_g), with regular input spike patterns 

```python
def train_pcm_trace( Input_spikes_train, F, iters = 250 ):
    nu = torch.nn.Parameter( torch.tensor(0.1).type(dtype=dtype) )
    delta_g = torch.nn.Parameter( torch.tensor((1-decay)/2).type(dtype=dtype) )
    params = [nu, delta_g]
    optimizer = torch.optim.SGD( params, lr = 1 )
    loss_hist = []
    for i in range(iters):
        G = pcm_convolve( Input_spikes_train, nu=nu, delta_g=delta_g, dt=time_step )
        MSE = torch.mean( torch.pow( (F-G),2 ) )
        optimizer.zero_grad()
        MSE.backward()
        optimizer.step()
        loss_hist.append( MSE.item() )
    return nu, delta_g, loss_hist
```

```python
freqs_input = [0.1e2,0.25e2,0.5e2,1e2,2e2]  # Hz
nus, delta_gs = [], []; 

for f, f_input in enumerate( freqs_input ):
    p_spike = f_input * time_step
    one_every = np.round( p_spike**-1,0 ).astype(int)
    spikes = np.arange( 0,seq_len, one_every )
    Input_spikes_train = torch.zeros(batches,seq_len,neurons).type(dtype=dtype)
    Input_spikes_train[:,spikes] = 1
    
    # Compute F
    F = exp_convolve( Input_spikes_train, decay )
    # Train PMC trace
    nu, delta_g, _ = train_pcm_trace(  Input_spikes_train, F, iters = 250 )
    nus.append(nu.item()); delta_gs.append( delta_g.item() )
```

```python
nus = np.array( nus )
delta_gs = np.array( delta_gs )
freqs_input = np.array(freqs_input)
```

```python
nus[-1], delta_gs[-1] = nu.item(), delta_g.item(); freqs_input[-1] = 1.2e2
```
