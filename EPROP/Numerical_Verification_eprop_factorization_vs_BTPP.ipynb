```python
# Copyright 2020, the e-prop team
# Full paper: A solution to the learning dilemma for recurrent networks of spiking neurons
# Authors: G Bellec*, F Scherr*, A Subramoney, E Hajek, Darjan Salaj, R Legenstein, W Maass
#
# In this check we verify numerically the exactness of the equation (1) of the paper.
# In other word, that the loss gradient dE/dW_ji can be computed with the formula:
#
# dE/dW_ji = sum_t dE/dz_j(t) [d z/dW_ji (t) ]_local
#
# As a ground it will be computed to the gradient computed with BPTT via auto-differentiation.
# This numerical verification is contained in a single script and follows the structure:
#
# 1. Let's define some parameters
# 2. Define the network model and the inputs
# 3. Simulate the network
# 4. Compute the true learning signal: dE/dz (total derivative) for an arbitrary loss function
# 5. Compute the eligibility traces for ALIFs using formula (25)
# 6. Compute the gradients with equation (1)
# 7. Compute the gradients given by BPTT using auto-diff
# 8. Start the tensorflow session and compute numerical verification.
#
# The relative difference between the two resulting gradients dE/dW_ij are approximately 10^-14.
# This tiny difference is the expected machine precision for two different computation schemes of the same gradient.
#
# This script requires was tested with tensorflow 1.15 and python3.6.
# More details requirements are explained in the folder Figure_3_and_S7_...
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False) ## for Dark plots
```

```python
from models import EligALIF, exp_convolve, shift_by_one_time_step
from tools import raster_plot
```

```python
# 1. Let's define some parameters
n_in = 3
n_LIF = 4
n_ALIF = 4
n_rec = n_ALIF + n_LIF

dt = 1  # ms
tau_v = 20  # ms
tau_a = 500  # ms
T = 50  # ms
f0 = 100  # Hz

thr = 0.62
beta = 0.07 * torch.cat( (torch.zeros(n_LIF,dtype=torch.float32),torch.ones(n_ALIF,dtype=torch.float32)), dim=0)
dampening_factor = 0.3
n_ref = 2
```

```python
# 2. Define the network model and the inputs
cell = EligALIF(n_in=n_in, n_rec=n_LIF + n_ALIF, tau=tau_v, beta=beta, thr=thr,
                dt=dt, tau_adaptation=tau_a, dampening_factor=dampening_factor,
                stop_z_gradients=False,  # here it computes the BPTT gradients, set it to True to compute instead e-prop with auto-diff
                n_refractory=n_ref)

np.random.seed(14)
#inputs_np = np.random.rand( 1,T,n_in ) < f0 * dt / 1000
#inputs = torch.from_numpy( inputs_np ).type(dtype=torch.float32)
inputs = ( torch.rand(1,T,n_in, dtype=torch.float32) < f0 * dt / 1000 ).type(dtype=torch.float32)
```

```python
file_load_init = r'C:\Users\FM261720\Desktop\Jupyter_Notebooks\EPROP\initialization.npz'
data = np.load( file_load_init )
w_rec_val = data['w_rec_val']
w_in_val  = data['w_in_val']
y_target_np = data['y_target_np']
w_out_np = data['w_out_np']
```

```python
cell.w_rec_val = torch.nn.Parameter( torch.from_numpy(w_rec_val).type(dtype=torch.float32) )
cell.w_in_var  = torch.nn.Parameter( torch.from_numpy(w_in_val).type(dtype=torch.float32) )
```

```python
# 3. Simulate the network,
# Using a for loop instead of using tf.nn.dynamic_rnn(...) is not efficient
# belows allows us to compute the true learning signals: dE/dz (total derivative)
# with auto-diff to perform the numerical verification
spikes = []
hidden_states = []
state = cell.zero_state(1,torch.float32) # ,n_rec=n_rec
for t in range(T):
    outs, state = cell(inputs[:,t],state)
    spikes.append(outs[0])
    hidden_states.append(outs[1])
```

```python
# 4. Compute the true learning signal: dE/dz (total derivative) for an arbitrary loss function
# (here a regression with a random signal)
#w_out = torch.randn(n_rec,1)
w_out = torch.nn.Parameter( torch.from_numpy( w_out_np ).type(dtype=torch.float32) )
decay_out = torch.exp( torch.tensor(-1/20).type(torch.float32) )
#decay_out = np_tensors['decay_out']
z_filtered = exp_convolve(torch.stack(spikes,1),decay_out)
y_out = torch.matmul(z_filtered,w_out)
#y_target = torch.rand(1,T,1)
y_target=torch.from_numpy( y_target_np ).type(dtype=torch.float32)
loss = torch.mean(torch.square(y_out - y_target))
```

```python
# This defines the true learning signal: dE/dz (total derivative)
#dE_dz = tf.gradients(loss, [spikes[t] for t in range(T)])
dE_dz = torch.autograd.grad(loss,  [spikes[t] for t in range(T)], retain_graph =True)
```

```python
# Stack the lists as tensors (second dimension is time)
# - spikes and learning signals will have shape: [n_batch, n_time , n_neuron]
# - eligibility traces will have shape: [n_batch, n_time , n_neuron, n_neuron]
spikes = torch.stack(spikes,dim=1)
hidden_states = torch.stack(hidden_states,dim=1)
dE_dz = torch.stack(dE_dz,dim=1)
```

```python
# 5. Compute the eligibility traces for ALIFs using formula (25)
v_scaled = cell.compute_v_relative_to_threshold_values(hidden_states)
spikes_last_time_step = shift_by_one_time_step(spikes)
eligibility_traces, epsilon_v, epsilon_a, psi = cell.compute_eligibility_traces(v_scaled, spikes_last_time_step, spikes, True)
```

```python
eligibility_traces_pcm, epsilon_v_pcm, epsilon_a_pcm, psi_pcm = cell.compute_eligibility_traces_pcm(v_scaled, spikes_last_time_step, spikes, True)
```

```python
# 6. Compute the gradients with equation (1)
gradients_eprop = torch.einsum('btj,btij->ij',dE_dz,eligibility_traces)
gradients_eprop_pcm = torch.einsum('btj,btij->ij',dE_dz,eligibility_traces_pcm)
```

```python
# 7. Compute the gradients given by BPTT using auto-diff
gradients_BPTT = torch.autograd.grad(loss, cell.w_rec_val, retain_graph=True)[0]
gradients_BPTT_out = torch.autograd.grad(loss, w_out, retain_graph=True)[0]
```

```python
# Show the gradients
fig, ax_list = plt.subplots(1,3, figsize=(15,7))
ax_list[2].imshow(gradients_eprop_pcm.detach().numpy())
ax_list[2].set_title("Gradient dE/dW_ji with pcm-prop", size=20)
ax_list[1].imshow(gradients_eprop.detach().numpy())
ax_list[1].set_title("Gradient dE/dW_ji with e-prop", size=20)
ax_list[0].imshow(gradients_BPTT.detach().numpy())
ax_list[0].set_title("Gradient dE/dW_ji with BPTT", size=20)

# Compute the relative error:
g_pcm_prop = gradients_eprop_pcm.clone().detach().numpy()
g_e_prop = gradients_eprop.clone().detach().numpy()
g_bptt = gradients_BPTT.clone().detach().numpy()
M = np.max(np.abs(g_bptt))
g_pcm_prop /= M
g_e_prop /= M
g_bptt /= M

gradient_errors = (g_e_prop - g_bptt)**2
gradient_errors_pcm = (g_pcm_prop - g_bptt)**2
max_gradient_errors = np.max(gradient_errors)
max_gradient_errors_pcm = np.max(gradient_errors_pcm)
print("Gradients computed with BPTT and auto-differentiation:")
print(np.array_str(gradients_BPTT.detach().numpy(), precision=3,suppress_small=True))
print("Gradients computed with the e-prop factorization (equation (1) in the paper):")
print(np.array_str(gradients_eprop.detach().numpy(), precision=3,suppress_small=True))
print("Gradients computed with the pcm-prop factorization:")
print(np.array_str(gradients_eprop_pcm.detach().numpy(), precision=3,suppress_small=True))
print("BPTT vs Eprop    : maximum element wise errors: {}".format(max_gradient_errors))
print("BPTT vs PCMPprop : maximum element wise errors: {}".format(max_gradient_errors_pcm))

fig, ax_list = plt.subplots(4, figsize=(8, 12), sharex=True)
raster_plot(ax_list[0],inputs.detach().numpy()[0], color='orange')
ax_list[0].set_ylabel("Input spikes")

raster_plot(ax_list[1],spikes.detach().numpy()[0], color='b')
ax_list[1].set_ylabel("Spikes")

v_max = np.max(np.abs(dE_dz.detach().numpy()))
ax_list[2].pcolor(np.arange(T),np.arange(n_rec),dE_dz.detach().numpy()[0].T,cmap='seismic',vmin=-1,vmax=1)
ax_list[2].set_ylabel("Learning signals")

for i in range(3):
    for j in range(3):
        if i != j:
            ax_list[3].plot(np.arange(T),eligibility_traces.detach().numpy()[0,:,i,j])
ax_list[3].set_ylabel("Eligibility traces")

plt.show()
```

```python
### VECTORIZED VERSION: TO BE DEVELOPED ###
z_pre = spikes_last_time_step.permute(1,0,2)
eps = torch.zeros( 50, 1, 8, 8 )
eps_ref = torch.zeros( 1, 8 )
Tevolve = torch.zeros( 1, 8 )
nu = 0.15
for t in range( z_pre.size(0) ):
    b_spk, idx_spk  = torch.where( z_pre[t] == 1 )
    b_notspk, idx_notspk = torch.where( z_pre[t] != 1 )
    Tevolve[b_spk, idx_spk] = 0.0
    Tevolve[b_notspk, idx_notspk] = Tevolve[b_notspk, idx_notspk] + cell.dt
    eps_ref[b_spk, idx_spk] = eps[t-1, b_spk, 0, idx_spk] + z_pre[t, b_spk, idx_spk]
    eps[t] = eps_ref[:,:,None].repeat( 1,1,eps_ref.size(-1) ) * ( cell.dt/( cell.dt + Tevolve ) )**nu
```

```python
### NON VECTORIZED VERSION ###
z_pre = spikes_last_time_step.permute(1,0,2)
eps = torch.zeros( 50, 1, 8, 8 )
eps_ref = torch.zeros( 1, 8 )
Tevolve = torch.zeros( 1, 8 )
nu = 0.25
for i in range( 8 ):
    for j in range( 8 ):
        Tevolve[:,i] = 0.0
        eps_ref[:,i] = 0.0
        for t in range( z_pre.size(0) ):
            spk = torch.where( z_pre[t,:,i] == 1 )[0]
            notspk = torch.where( z_pre[t,:,i] != 1 )[0]
            Tevolve[spk, i] = 0.0
            Tevolve[notspk, i] = Tevolve[notspk, i] + cell.dt
            eps_ref[spk, i] = torch.add( eps[t-1, spk, i, j], 1 )
            eps[t,:,i,j] = eps_ref[:, i] * ( cell.dt/( cell.dt + Tevolve[:, i] ) )**nu
```
